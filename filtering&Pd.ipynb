# Fixing the 'false' value in the column '1990' and the dtype of that column

```python
ghg[ghg['Country/Region'] == 'Namibia']#['1990']
```

```python
mean1990 = ghg[ghg['1990'] != 'false']['1990'].astype('float64').mean()
print(mean1990)
#ghg[ghg['1990'] == 'false']
ghg[ghg['1990'] == 'false']['1990'] = mean1990
ghg.loc[ghg['1990'] == 'false', '1990'] = mean1990
#print(ghg[ghg['1990'] == mean1990])
ghg['1990'] = ghg['1990'].astype('float64')
ghg[ghg['Country/Region'] == "Namibia"]
ghg[ghg['Country/Region'] == "Namibia"]['1990']
#ghg.info()
#ghg.head()
```
