# dtype

```python
arr = np.array([i for i in range(10)])
print('arr:', arr)
print('arr.dtype:', arr.dtype)
print('---')
floatArr = np.array([i for i in range(10)], dtype = 'float64')
print('floatArr:', floatArr)
print('floatArr.dtype:', floatArr.dtype)
print('---')
arr = arr.astype(np.float64)
print('arr:', arr)
print('arr.dtype:', arr.dtype)
print('---')
arr = arr.astype(np.bool_)
print('arr:', arr)
print('arr.dtype:', arr.dtype)
```

# ndim

```python
import numpy as np
```

```python
scalar = np.array(22)
print('scalar: \t', scalar)
print('scalar.ndim: \t', scalar.ndim)

matrix = np.array([[1, 3, 5], [4, 8, 7]])
print('matrix:')
print(matrix)
print('matrix.ndim: \t', matrix.ndim)
```

# Add or eliminate dimensions

```python
vec = np.array([1, 2, 3], ndmin = 10)
print(vec)
print(vec.ndim)
```

```python
expanded = np.expand_dims( np.array([1, 2, 3]), axis = 0 )
print('expanded: ',  expanded)
print('expanded.ndim', expanded.ndim)
```

```python
expanded = np.expand_dims( np.array([1, 2, 3]), axis = 1 )
print('expanded: ',  expanded)
print('expanded.ndim', expanded.ndim)
```

```python
print(vec, vec.ndim)
vec2 = np.squeeze(vec)
print(vec2, vec2.ndim)
```
