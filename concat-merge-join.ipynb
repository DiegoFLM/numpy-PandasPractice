# pd.merge

```python
left = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0','B1','B2','B3']
})

right = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
    'C':['C0', 'C1', 'C2', 'C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
#left.merge(right) 
#ow
left.merge(right, on = 'key')
```

```python
left = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0','B1','B2','B3']
})

right = pd.DataFrame({'key2':['k0', 'k1', 'k2', 'k3'],
    'C':['C0', 'C1', 'C2', 'C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
left.merge(right, left_on = 'key', right_on = 'key2')
```

```python
left = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0','B1','B2','B3']
})

right = pd.DataFrame({'key2':['k0', 'k1', 'k2', 'k8'],
    'C':['C0', 'C1', 'C2', 'C8'],
    'D':['D0','D1','D2','D8']
})
```

```python
left.merge(right, left_on = 'key', right_on = 'key2', how = 'outer')
```

# pd.join: match by index

```python
left = pd.DataFrame({'A':['A0', 'A1', 'A2'],
    'B':['B0','B1','B2']
}, index = ['k0', 'k1', 'k2'])

right = pd.DataFrame({'C':['C0', 'C1', 'C2'],
    'D':['D0','D1','D2']
}, index = ['k0', 'k2', 'k3'])

left
```

```python
left.join(right, how = 'outer')
```

# pivot

```python
df_books = pd.read_csv('files/bestsellers-with-categories.csv',sep=',',header=0)
df_books.head()
```

```python
df_books.info()
```

```python
pivotTable = df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
pivotTable.head()
```

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```
