```python
import numpy as np
import pandas as pd
```

```python
ghg = pd.read_csv('files/ghg-emissions.csv', nrows = 193) 
ghg.head()
```

### Fixing 'false' value

```python
mean1990 = ghg[ghg['1990'] != 'false']['1990'].astype('float64').mean()
print(mean1990)
#ghg[ghg['1990'] == 'false']['1990'] = mean1990 #doesn't work. The slice is a copy, not a reference.
ghg.loc[ghg['1990'] == 'false', '1990'] = mean1990
ghg['1990'] = ghg['1990'].astype('float64')
ghg[ghg['Country/Region'] == 'Namibia']#['1990']
```

# bestSellers

```python
dfBestSellers = pd.read_csv('files/bestsellers-with-categories.csv')
dfBestSellers.head()
```

```python
dfBestSellers.groupby('Author').count() #.mean() #.sum()
```

```python
dfBestSellers.groupby('Author').sum().loc['William Davis']
```

```python
dfBestSellers.groupby('Author').sum().reset_index()
```

```python
dfBestSellers.groupby('Author').agg(['min', 'max'])
```

```python
dfBestSellers.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating': 'sum'})
```

```python
dfBestSellers.groupby(['Author', 'Year']).count()
```

## lambda functions can be used in agg() for calculating values

```python
dfBestSellers.groupby('Author').agg({'Year':lambda x: [i-2000 for i in x], 'Reviews':lambda x: sum([i**2 for i in x])})
```
